```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

def df(x, y):
    result = list()
    n = len(x)
    h = x[1] - x[0]
    for i in range(n):
        if i == 0:
            result.append((y[i + 1] - y[i]) / h)
        elif i + 1 == n:
            result.append((y[i] - y[i - 1]) / h)
        else:
            result.append((y[i + 1] - y[i - 1]) / (2 * h))
    return result
            
def ddf(x, y):
    result = list()
    n = len(x)
    h = x[1] - x[0]
    for i in range(n):
        if i == 0:
            result.append((y[i + 2] - 2 * y[i + 1] + y[i]) / (h * h))
        elif i + 1 == n:
            result.append((y[i] - 2 * y[i - 1] + y[i - 2]) / (h * h))
        else:
            result.append((y[i + 1] - 2 * y[i] + y[i - 1]) / (h * h))
    return result

def y1(x, a):
    summ = 0
    for i in range(len(a)):
        summ += (a[i] * (x**i))
    return summ    

def phi3(x, a):
    result = list()
    for i in x:
        result.append(y1(i, a))
    return result
    
x2 = np.arange(0.3, 1.8, 0.1)
y2 = phi3(x2, [3.81000121,  2.30779827, -1.37962274,  0.28806357])
dy2 = df(x2, y2)
ddy2 = ddf(x2, y2)

x = np.array([0.3, 0.6, 0.9, 1.2, 1.5, 1.8])
y = np.array([4.39, 4.75, 4.98, 5.11, 5.12, 5.18])
df1 = df(x, y)
ddf1 = ddf(x, y)

fig, axs = plt.subplots(5, figsize=tuple((12.8, 25.6)), constrained_layout=True)
fig.suptitle("diff", fontsize=28.0)

axs[0].set_title('f', fontsize=14)
axs[1].set_title('df, ddf', fontsize=14)
axs[2].set_title('f*', fontsize=14)
axs[3].set_title('df*', fontsize=14)
axs[4].set_title('ddf*', fontsize=14)

axs[0].plot(x, y, 'r')
axs[0].plot(x, y, 'b.')
axs[1].plot(x, df1, 'g.')
axs[1].plot(x, ddf1, 'b.')

axs[2].plot(x2, y2, 'r')
axs[3].plot(x2, dy2, 'c')
axs[4].plot(x2, ddy2, 'm')

for i in range(len(x)):
    axs[0].plot(np.array([x[i] - 0.12, x[i] + 0.12]), np.array([df1[i] * (-0.12) + y[i], df1[i] * (0.12) + y[i]]), 'g')
    # axs[1].plot(np.array([x[i] - 0.12, x[i] + 0.12]), np.array([ddf1[i] * (0.12) + df1[i], ddf1[i] * (-0.12) + df1[i]]), 'g')

axs[0].set(xlim=(0.1, 1.9))
axs[1].set(xlim=(0.1, 1.9))
axs[2].set(xlim=(0.1, 1.9))
axs[3].set(xlim=(0.1, 1.9))
axs[4].set(xlim=(0.1, 1.9))

axs[0].grid(True)
axs[1].grid(True)
axs[2].grid(True)
axs[3].grid(True)
axs[4].grid(True)

plt.show()
```
